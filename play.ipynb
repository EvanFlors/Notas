```python
def gainMaxValue(board: list[int], k: int) -> int:
    n = len(board)
    if not board:
        return 0

    dp = [0] * k
    max_total = float('-inf')

    for i in range(n - 1, -1, -1):
        current_score = board[i]
        if i + k < n:
            current_score += dp[i % k]

        dp[i % k] = current_score
        if current_score > max_total:
            max_total = current_score

    return max_total

if __name__ == '__main__':
    game_val = [2, -3, 4, 6, 1]
    k = 2
    result = gainMaxValue(game_val, k)

    print(result)
```

```python
def getMaxValueSum(encrypted_file, binary, k):
    n = len(encrypted_file)

    sum_ones = sum(val for val, b in zip(encrypted_file, binary) if b == 1)

    prefix = [0] * (n + 1)
    for i in range(n):
        prefix[i + 1] = prefix[i] + (encrypted_file[i] if binary[i] == 0 else 0)

    max_window = max(prefix[i + k] - prefix[i] for i in range(n - k + 1))

    return sum_ones + max_window

if __name__ == '__main__':
    result = getMaxValueSum([7, 4, 3, 5], [1, 0, 0, 0], 2)
    print(result)
```
